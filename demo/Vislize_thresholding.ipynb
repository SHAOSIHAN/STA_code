# 可视化阈值

```python
# stage one
y = stage1_1
x = np.linspace(0, len(y)-1, len(y))
avg = y.mean()
plt.plot(x, y, label='thresholding')
plt.legend()
plt.axhline(y=avg,ls=":",c="red")
plt.savefig("stage1_1.png",
            dpi=800,
            bbox_inches = 'tight')#解决图片不清晰，不完整的问题
plt.show()
y.mean()
```
